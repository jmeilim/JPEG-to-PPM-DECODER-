## 1. Introduction

On souhaite décrire l'évolution de deux populations en interaction: une population de proies
(par exemple des lièvres)
et une population de prédateurs (par exemple des lynx).  On note respectivement $x(t)$ et $y(t)$ les effectifs des
proies et prédateurs 
(normalisés par des valeurs de référence) à l'instant $t$.
On peut modéliser l'évolution de ces populations
à l'aide d'équations différentielles de la forme:
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) \, f(x(t),y(t),t) \\
y'(t) &=& y(t) \, g(x(t),y(t),t)
\end{array}
\right.
\end{equation}
où $f,g$ sont des fonctions $C^1$ qui décrivent les taux de croissance de chaque population.
La fonction $y\mapsto f(x,y,t)$ est décroissante (terme de prédation) et
$x\mapsto g(x,y,t)$ est croissante.
Par ailleurs, on suppose les fonctions 
$x\mapsto f(x,y,t)$ et $y\mapsto g(x,y,t)$ décroissantes
si on veut prendre en compte une quantité limitée de nourriture.
La dépendance de $f,g$ par rapport au temps permet
de tenir compte des variations saisonnières
de la quantité de nourriture et des taux de natalité.

Dans ce TP, nous allons étudier des modèles de ce type pour des choix particuliers de fonctions $f$ et $g$
décrits dans les sections 2 et 3.

## 2. Equation de Lotka-Volterra

Il s'agit d'une version très simplifiée du modèle précédent, dans laquelle 
la population de proies dispose de ressources illimitées et 
l'équation différentielle est autonome, i.e. les fonctions $f,g$ ne
dépendent pas de $t$: 
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) ( 1- y(t) ) \\
y'(t) &=& \alpha\, y(t)(x(t)  - 1).
\end{array}
\right.
\end{equation}
Le paramètre $\alpha$ est strictement positif.
Pour les calculs numériques, on fixera $\alpha =1$.

### 2.1. Analyse du modèle

$$

Sans prédateur, la population des proies croît à l'infini.

$$

#### Question 4
Une fonction $H\, : \, ]0,+\infty[ \times ]0,+\infty[ \rightarrow \mathbb{R}$
est appelée intégrale première de l'équation différentielle
lorsque pour toute solution $(x(t),y(t))$ on a $\frac{d}{dt} H(x(t),y(t))=0$.
Vérifier que 
$$
H(x,y)=\alpha\, (x-\ln{x})+y-\ln{y}
$$
est une intégrale première.

$$
\begin{align}
\frac{d}{dt} H(x(t),y(t)) &= \frac{d}{dt} \left( \alpha\, (x(t)-\ln{x(t)})+y(t)-\ln{y(t)} \right) \\
                          &= \alpha x'(t) (1 - \frac{1}{x(t)}) + y'(t) (1 - \frac{1}{y(t)}) \\
                          &= \alpha x(t)(1-y(t)) (1 - \frac{1}{x(t)}) + \alpha y(t)(x(t) - 1) (1 - \frac{1}{y(t)}) \\
                          &= \alpha (x(t)-1)(1-y(t)) + \alpha (x(t)-1)(y(t)-1) \\
                          &= \alpha (x(t)-1)(1-y(t)+y(t)-1) \\
\frac{d}{dt} H(x(t),y(t)) &= 0
\end{align}
$$

#### Question 5
Tracer différentes courbes de niveau 
$H(x,y)=\mbox{constante}$ (matplotlib.pyplot.contour)
ainsi que le champ de vecteurs $F(x,y)=(\, x ( 1- y )\, ,\, \alpha\, y(x  - 1)\, )$ 
du système (matplotlib.pyplot.quiver) pour $\alpha =1$
(on pourra normaliser $F$ pour plus de lisibilité).

```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
def H(x, y, alpha):
    return alpha * (x - np.log(x)) + y - np.log(y)

def F(x, y, alpha):
    dxdt = x * (1 - y)
    dydt = alpha * y * (x - 1)
    return dxdt, dydt

alpha = 1
x_range = np.linspace(0.01, 4, 30)  
y_range = np.linspace(0.01, 4, 30)  

X, Y = np.meshgrid(x_range, y_range)
Z = H(X, Y, alpha)

plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, levels=30, cmap='jet')

U, V = F(X, Y, alpha)

plt.quiver(X, Y, U, V, color='black', scale=80)

plt.title(f'Courbes de niveau de H(x, y) et champ de vecteurs F(x, y) pour alpha = {alpha}')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0, 4])
plt.ylim([0, 4])
plt.show()
```

### 2.2. Intégration Numérique

On discrétise maintenant le système de Lotka-Volterra pour $\alpha =1$ 
en utilisant différents schémas numériques.

#### 2.2.1. Schéma d'Euler explicite
Afin d'approcher numériquement les solutions du modèle de Lotka-Volterra
pour $t \in [0,t_{\mathrm{max}}]$, 
on discrétise l'équation à l'aide du schéma d'Euler explicite:
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}&=& x_k + h\, x_k ( 1- y_k ) \\
y_{k+1} &=& y_k + h\, y_k (x_k  - 1)
\end{array}
\right.
\end{equation}
où $h >0$ est le pas d'intégration en temps et
$(x_k,y_k)$ une approximation de $(x(t_k),y (t_k))$ en $t_k = k\, h$.
On fixe $h=t_{\mathrm{max}}/N$ où $N$ est un entier $\geq 1$.

#### Question 7
Pour $t_{\mathrm{max}}=120$, $(x(0),y(0))=(2,1)$ 
et différentes valeurs du pas de temps $h$, 
tracer les trajectoires 
$(x_k,y_k)_{0 \leq k \leq N} \subset {\left(\mathbb{R}^+ \right)}^2$
ainsi que les graphes de $x_k$, $y_k$ en fonction de $t_k$. 
Qu'observez-vous ? Comparer ces résultats à l'étude théorique 
de la question 6.

```python
tmax=120
for h in [0.01 , 0.05  ]:
    N = int(tmax/h)
    X=[2]
    Y=[1]
    for i in range(1,N):
        xi=X[i-1] + h*X[i-1]*(1-Y[i-1])
        yi=Y[i-1] + h*Y[i-1]*(X[i-1]-1)
        X.append(xi)
        Y.append(yi)
    t = [i * h for i in range(N)] 
    
    plt.figure(figsize=(10, 6))
    plt.plot(X, Y, label="Trajectoire (x_k, y_k)")
    plt.xlabel("Proies (x_k)")
    plt.ylabel("Prédateurs (y_k)")
    plt.title("Trajectoires dans le plan des phases")
    
    
    plt.show()
    
    plt.figure(figsize=(10, 6))
    plt.plot(t, X, label="Proies (x)" )
    plt.plot(t, Y, label="Prédateurs (y)")
    plt.xlabel("Temps (t)")
    plt.ylabel("Population")
    plt.title("Trajectoires des populations de proies et prédateurs")
    
    
    plt.show()
```

Les trajectoires de $(x_k,y_k)_{0 \leq k \leq N} \subset {\left(\mathbb{R}^+ \right)}^2$ sont sous forme de spirale sortante, autrement dit elles ne suivent pas les courbes de niveau définies par $ (H(x(t), y(t)) = \text{constante}) $ et elles divergent (les proies et les prédateurs sont en croissance infinie).
Les fonctions $t \mapsto (x(t),y(t))$ oscillent périodiquement dans le temps mais avec divergence des amplitudes, présentant donc une instabilité numérique.

#### 2.2.2. Schéma d'Euler implicite
On considère maintenant le schéma d'Euler implicite
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}-x_k&=&   h\, x_{k+1} ( 1- y_{k+1} ) \\
y_{k+1}-y_k &=&   h\, y_{k+1} (x_{k+1}  - 1)
\end{array}
\right.
\end{equation}

#### Question 8
Exprimer $x_{k+1}$ en fonction de $x_k$, $y_{k+1}$ et $h$,
puis calculer $y_{k+1}$ en fonction de $(x_k,y_{k},h)$
en résolvant une équation du second degré.

Remarque: lorsque $h=0$ on doit obtenir $(x_{k+1},y_{k+1})=(x_{k},y_{k})$.


1. **Première équation :**
   $$
   x_{k+1} - x_k = h \cdot x_{k+1} \cdot (1 - y_{k+1})
   $$
   En réarrangeant, on obtient :
   $$
   x_{k+1} \cdot (1 - h \cdot (1 - y_{k+1})) = x_k
   $$
   Donc :
   $$
   x_{k+1} = \frac{x_k}{1 - h \cdot (1 - y_{k+1})}
   $$

2. **Deuxième équation :**
   $$
   y_{k+1} - y_k = h \cdot y_{k+1} \cdot (x_{k+1} - 1)
   $$
   En réarrangeant, on obtient :
   $$
   y_{k+1} \cdot (1 - h \cdot (x_{k+1} - 1)) = y_k
   $$
   Donc :
   $$
   y_{k+1} = \frac{y_k}{1 - h \cdot (x_{k+1} - 1)}
   $$

3. **Substitution de $(x_{k+1})$ dans $(y_{k+1})$ :**
   $$
   y_{k+1} = \frac{y_k}{1 - h \cdot \left(\frac{x_k}{1 - h \cdot (1 - y_{k+1})} - 1\right)}
   $$

4. **Simplification pour obtenir une équation du second degré :**
   Après simplification, on obtient une équation du second degré pour $(y_{k+1})$ :
   $$
   A \cdot y_{k+1}^2 + B \cdot y_{k+1} + C = 0
   $$
   où :
   $$
   A = h(h+1) 
   $$
   $$
   B = (1 - h^2 - h \cdot x_k - h \cdot y_k)
   $$
   $$
   C = y_k \cdot (h - 1)
   $$

5. **Solution de l'équation du second degré :**
   La solution pour $(y_{k+1})$ est donnée par la formule  :
   $$
   y_{k+1} = \frac{-B + \sqrt{B^2 - 4 \cdot A \cdot C}}{2 \cdot A}
   $$
   (On choisit la solution positive car $(y_{k+1})$ représente une population.)
   En explicitant (A), (B), et (C) :
$$[ y_{k+1} = \frac{-(1 - h^2 - h \cdot x_k - h \cdot y_k) + \sqrt{(1 - h^2 - h \cdot x_k - h \cdot y_k)^2 - 4 \cdot h \cdot (h + 1) \cdot x_k \cdot y_k \cdot (h - 1)}}{2 \cdot h \cdot (h + 1) \cdot x_k} ]$$


#### Question 9
Reprendre la question 7 pour le schéma d'Euler implicite.

```python

tmax=120
for h in [0.01 , 0.05  ]:
    N = int(tmax/h)
    X = [2] 
    Y = [1]  


    for i in range(1, N):
        x_k = X[i-1]
        y_k = Y[i-1]
        
    
        A = h * (h + 1) 
        B = 1 - h**2 - h * x_k - h * y_k
        C = y_k * (h - 1)
        
        
        discriminant = B**2 - 4 * A * C
        if discriminant < 0:
            raise ValueError(f"Discriminant négatif à l'étape {i}, instabilité numérique détectée.")
        
        y_next = (-B + np.sqrt(discriminant)) / (2 * A) 
    
        x_next = x_k / (1 - h * (1 - y_next))
        
    
        X.append(x_next)
        Y.append(y_next)

    t = [i * h for i in range(N)]


    plt.figure(figsize=(10, 6))
    plt.plot(X, Y, label="Trajectoire (x_k, y_k)")
    plt.xlabel("Proies (x_k)")
    plt.ylabel("Prédateurs (y_k)")
    plt.title("Trajectoires dans le plan des phases")
    plt.legend()
    plt.grid()
    plt.show()


    plt.figure(figsize=(10, 6))
    plt.plot(t, X, label="Proies (x)")
    plt.plot(t, Y, label="Prédateurs (y)")
    plt.xlabel("Temps (t)")
    plt.ylabel("Population")
    plt.title("Trajectoires des populations de proies et prédateurs")
    plt.legend()
    plt.grid()
    plt.show()
```

Les trajectoires de $(x_k,y_k)_{0 \leq k \leq N} \subset {\left(\mathbb{R}^+ \right)}^2$ sont sous forme de spirale entrante, autrement dit elles ne suivent pas les courbes de niveau définies par $ (H(x(t), y(t)) = \text{constante}) $ et elles convergent vers le point d'équilibre (1, 1) (schéma stable).
Les fonctions $t \mapsto (x(t),y(t))$ oscillent périodiquement dans le temps mais avec convergence des amplitudes.

#### 2.2.3. Schéma d'Euler semi-implicite
On considère un schéma d'Euler semi-implicite
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}-x_k&=&  h\, x_{k} ( 1- y_{k+1} ) \\
y_{k+1} -y_k&=&  h\, y_{k+1} (x_{k}  - 1)
\end{array}
\right.
\end{equation}
dans lequel la variable $x$ est traitée de manière explicite
(évaluée en $t=t_{k}$ dans le second membre de l'équation différentielle)
et $y$ de manière implicite (évaluée en $t=t_{k+1}$).

Les fonctions $t \mapsto (x(t),y(t))$ oscillent périodiquement dans le temps avec une amplitude constante, ce qui implique une compensation continue entre les taux des proies et des prédateurs.

## 3. Modèle proies-prédateurs avec croissance limitée
On considère l'équation différentielle:
\begin{equation}
\left\{
\begin{array}{lll}
x'(t)&=& x(t) ( 1- y(t) - c(t)\, x(t)) \\
y'(t) &=&  y(t)(x(t)  - 1)
\end{array}
\right.
\end{equation}
où le coefficient $c(t)>0$ 
tient compte du fait que la population de proies est limitée par la quantité de nourriture.
On suppose que la fonction $c$ est $T$-périodique
(variations saisonnières des ressources) avec $T=12$ (l'unité de temps représente un mois).
Dans ce TP on fixera 
$$
c(t)=\lambda\, \left(1+\frac{1}{2}\, \cos{\left( \frac{\pi}{6}\, t \right)}\right),
$$
où $\lambda >0$ est un paramètre.

On discrétise l'équation différentielle par le
schéma d'Euler semi-implicite
\begin{equation}
\left\{
\begin{array}{lll}
x_{k+1}-x_k&=&  h\, x_{k} ( 1- y_{k+1} - c(t_k)\, x_k) \\
y_{k+1} -y_k&=&  h\, y_{k+1} (x_{k}  - 1)
\end{array}
\right.
\end{equation}

#### Question 12
Exprimer  $(x_{k+1},y_{k+1})$ en fonction de $(t_k,x_k,y_{k},h)$.

$$ y_{k+1} = \frac{y_k}{1 - h \cdot (x_k - 1)}$$

$$ x_{k+1} = x_k + h \cdot x_k \cdot \left(1 - \frac{y_k}{1 - h \cdot (x_k - 1)} - c(t_k) \cdot x_k\right)$$
 où
  $$(c(t_k) = \lambda \cdot \left(1 + \frac{1}{2} \cos\left(\frac{\pi}{6} \cdot t_k\right)\right)).$$

#### Question 13
On fixe $t_{\mathrm{max}}=120$ et $h=10^{-3}$.
Pour différentes valeurs du paramètre $\lambda$,
tracer les graphes de $x_k$, $y_k$ en fonction de $t_k$
pour différentes conditions initiales $(x(0),y(0))$. 
Qu'observez-vous ? 
Interpréter le comportement des solutions lorsque
$t\rightarrow +\infty$
en terme d'évolution des populations, ainsi que l'effet du paramètre $\lambda$.

```python


t_max = 120
h = 0.001
N = int(t_max / h)
lambdas = [0.5, 1.0, 1.5]
initial_conditions = [
    (2, 1),
    (1.5, 0.5),
    (3, 2)
]

def c(t, lambd):
    return lambd * (1 + 0.5 * np.cos((np.pi / 6) * t))

for lambd in lambdas:
    plt.figure(figsize=(12, 6))
    for x0, y0 in initial_conditions:
        X = [x0]
        Y = [y0]
        t_values = [0]
        for i in range(1, N):
            
            t_k = i * h
            x_k = X[-1]
            y_k = Y[-1]
            y_next = y_k / (1 - h * (x_k - 1))
            x_next = x_k + h * x_k * (1 - y_next - c(t_k, lambd) * x_k)
            X.append(x_next)
            Y.append(y_next)
            t_values.append(t_k)
        plt.plot(t_values, X, label=f"Proies (x0={x0}, y0={y0})")
        plt.plot(t_values, Y, linestyle='--', label=f"Prédateurs (x0={x0}, y0={y0})")
    plt.title(f"Évolution des populations pour λ = {lambd}")
    plt.xlabel("Temps (t)")
    plt.ylabel("Population")
    plt.legend()
    plt.grid()
    plt.show()
```

Pour différentes valeurs du paramètre ( \lambda ), on observe que lorsque $ \lambda$  est faible, les oscillations des populations de proies et de prédateurs sont plus marquées et régulières, traduisant une faible limitation des ressources. À mesure que $\lambda$  augmente, les oscillations s'amortissent progressivement, et les populations convergent vers un point d'équilibre stable. Cela reflète l'effet de la limitation des ressources, qui réduit la capacité des proies à se reproduire, impactant indirectement les prédateurs. À long terme $( t \to +\infty )$, une augmentation de $ \lambda $ stabilise les populations, empêchant des fluctuations importantes et favorisant une coexistence équilibrée entre proies et prédateurs.
